```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
from google.colab import files
from IPython.display import display, Image
```

```python
# Upload an image
uploaded = files.upload()
image_path = list(uploaded.keys())[0]
```

```python
# Read the image
image = cv2.imread(image_path)
image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)  # Convert to RGB for proper display
```

```python
# Apply different blurring techniques
gaussian_blur = cv2.GaussianBlur(image, (15, 15), 0)
median_blur = cv2.medianBlur(image, 15)
bilateral_blur = cv2.bilateralFilter(image, 15, 75, 75)
```

```python
# Create an interactive UI using OpenCV
cv2.imwrite("original.jpg", cv2.cvtColor(image, cv2.COLOR_RGB2BGR))
cv2.imwrite("gaussian_blur.jpg", cv2.cvtColor(gaussian_blur, cv2.COLOR_RGB2BGR))
cv2.imwrite("median_blur.jpg", cv2.cvtColor(median_blur, cv2.COLOR_RGB2BGR))
cv2.imwrite("bilateral_blur.jpg", cv2.cvtColor(bilateral_blur, cv2.COLOR_RGB2BGR))

def display_images():
    fig, axes = plt.subplots(2, 2, figsize=(20, 10))
    images = [image, gaussian_blur, median_blur, bilateral_blur]
    titles = ["Original Image", "Gaussian Blur", "Median Blur", "Bilateral Filter"]

    for ax, img, title in zip(axes.flat, images, titles):
        ax.imshow(img)
        ax.set_title(title)
        ax.axis("off")

    plt.show()

display_images()

print("Download processed images:")
display(Image("original.jpg"))
display(Image("gaussian_blur.jpg"))
display(Image("median_blur.jpg"))
display(Image("bilateral_blur.jpg"))
```
